```python
import json

with open('JSONs/vagas_padrao.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
```

```python
import pandas as pd

df = pd.DataFrame.from_dict(data, orient='index')
display(df.head())
```

```python
import pandas as pd

df_normalized = pd.json_normalize(list(data.values()), sep='_')
display(df_normalized.head())
```

```python
print("Nomes das colunas:")
for col in df_normalized.columns:
    print(col)
```

```python
display(df_normalized['perfil_vaga_nivel profissional'].value_counts())
```

```python
#Stakeholder disse que 'Analista' é o mesmo que 'Júnior'

df_normalized['perfil_vaga_nivel profissional'] = df_normalized['perfil_vaga_nivel profissional'].str.replace('Analista', 'Júnior', regex=False)
```

```python
display(df_normalized['perfil_vaga_nivel profissional'].value_counts())
```

```python
total_vagas = len(df_normalized)
junior_vagas = df_normalized['perfil_vaga_nivel profissional'].str.lower().eq('júnior').sum()
porcentagem_junior = (junior_vagas / total_vagas) * 100
print(f"Porcentagem de vagas do tipo Júnior: {porcentagem_junior:.2f}%")

# Portanto, achamos que devemos atacar o mercado de vagas Júnior, que é o mais comum. As vagas Pleno e Sênior são menos comuns 
# e mais difíceis de lidar. Elas devem passar por um processo seletivo mais rigoroso, oque talvez exijam mais experiência do recrutador.
# Às máquinas, o mais simples e mais mecânico. Aos humanos, o mais complexo e subjetivo. HYBRID RESOURCERS.
```
